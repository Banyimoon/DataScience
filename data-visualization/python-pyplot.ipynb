```python
import datetime
import pandas as pd
import pandas.io.data

import numpy as np
import scipy


from matplotlib import pyplot as plt
import matplotlib.mlab as mlab

import plotly.plotly as py

%matplotlib inline
```

```python
x = np.arange(0, 2*np.pi, 0.05)
y1 = np.cos(x*2)
y2 = 2*np.exp(-x)*np.sin(x*5)

line1, line2 = plt.plot(x, y1, 'b-', x, y2, 'r-')
plt.setp(line1, color='b', linewidth=4)
plt.setp(line2, color='g', linewidth=4)

plt.axis([0, 2*np.pi, -1, 1.6])
plt.show()
```

```python
## Plotly version:

line1, line2 = plt.plot(x, y1, 'b-', x, y2, 'r-')
plt.setp(line1, color='b', linewidth=4)
plt.setp(line2, color='g', linewidth=4)

plt.axis([0, 2*np.pi, -1, 1.6])

mpl_fig1 = plt.gcf()

py.iplot_mpl(mpl_fig1, filename='DataScience/data-visualization/lineplot')
```
