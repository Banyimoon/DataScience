## 1. Line and Scatter Plots

### Line plot with trigonometric functions

```python
x = np.arange(0, 2*np.pi, 0.1)
y1 = np.cos(x*2)
y2 = 2*np.exp(-x/2)*np.sin(x*5)


fig1 = plt.figure(1, figsize=(8,6))
ax = fig1.add_subplot(1, 1, 1)
line1, line2 = plt.plot(x, y1, 'b-', x, y2, 'r-')
plt.setp(line1, color='b', linewidth=3, marker='^', markersize=8)
plt.setp(line2, color='g', linewidth=3, marker='o', markersize=8)

plt.setp(ax.get_xticklabels(), rotation='horizontal', fontsize=16)
plt.setp(ax.get_yticklabels(), rotation='vertical', fontsize=16)

plt.xlabel('X', size=20)
plt.ylabel('Y', size=20)
plt.title('Y vs X', size=20)
#plt.legend(['Y1', 'Y2'], loc='upper right') ## not supported by plotly yet

plt.axis([0, 2*np.pi, -1.4, 1.8])  
## or use : 
# plt.xlim([0, 2*np.pi])
# plt.ylim([-1.4, 1.8])


plt.show()
```

**Useful Parameters:**

  * linewidth (lw): float
  * linestyle (ls): '-'   '--'   ':'   '-.'
  * marker: string  'o'   '*'
  * markerfacecolor (mfc)
    

### Scatter plot with multivariate Gaussian distribution

```python
smat = np.array([[0.15, 0.08], [0.08, 0.1]])

d = np.random.multivariate_normal(mean=[4.0, 5.0], cov=smat, size=1000)

fig2 = plt.figure(2, figsize=(8,6))
ax2 = fig2.add_subplot(1, 1, 1)

plt.scatter(d[:,0], d[:,1], color='b', marker='o', s=80, alpha=0.25)
plt.xlabel('$x_1$', size=20)
plt.ylabel('$x_2$', size=20)
plt.title('Mutivariate Normal Distribution', size=20)
plt.xlim([2.5, 5.5])
plt.ylim([3.5, 6.5])
plt.text(2.8, 6.0, r'$\mathcal{N}(\bar\mu, \Sigma)$', fontsize=30)

plt.setp(ax2.get_xticklabels(), rotation='horizontal', fontsize=16)
plt.setp(ax2.get_yticklabels(), rotation='horizontal', fontsize=16)

plt.show()
```

## 2. Histograms

```python
d_norm_1 = np.random.normal(loc=0.0, scale=3.0, size=5000)

## Build a Gaussian Mixture Model:
array1 = np.random.normal(loc=4.0, scale=2.0, size=2000)
array2 = np.random.normal(loc=-5.0, scale=4.0, size=2000)
d_norm_2 = np.concatenate((array1, array2))

import prettytable as pt
table1 = pt.PrettyTable(["Data Index", "Mean", "Stan. Dev.", "Median"])
table1.align["Data Index"] = "c" # l:Left align c:centered
table1.padding_width = 1 # One space between column edges and contents (default)
table1.add_row([1, np.mean(d_norm_1), np.std(d_norm_1), np.median(d_norm_1)])
table1.add_row([2, np.mean(d_norm_2), np.std(d_norm_2), np.median(d_norm_2)])
print table1




fig3 = plt.figure(3, figsize=(8, 6))
ax3 = fig3.add_subplot(1, 1, 1)

plt.hist(d_norm_1, bins=40, normed=True, color='b', alpha=0.4)
plt.hist(d_norm_2, bins=40, normed=True, color='g', alpha=0.4)

plt.xlabel('$x$', size=20)
plt.ylabel('Probability Density', size=20)
plt.title('Histogram', size=20)

plt.setp(ax3.get_xticklabels(), rotation='horizontal', fontsize=16)
plt.setp(ax3.get_yticklabels(), rotation='horizontal', fontsize=16)

plt.show()
```
