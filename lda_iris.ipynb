```python
library(MASS)

data(iris)

print(head(iris))
```

```python
lda.iris <- lda(Species ~ ., iris, prior = c(1,1,1)/3)

lda.pred <- predict(lda.iris, iris)
```

```python
## Plot the data using LD1 & LD2 attributes

plot(lda.pred$x, pch=as.numeric(lda.pred$class)+16, cex=1.6, col=lda.pred$class)
```
